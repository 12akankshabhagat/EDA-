### how to get categorical col?


```python
cat_col = [col for col in df.columns if df[col].dtype == 'object']
cat_col 
```

### how to get numerical columns?

```python

num_col = [col for col in df.columns if df[col].dtype != 'object']
num_col
```

```python
for column in cat_col:
    plt.figure(figsize = (8, 5))
    sns.countplot(x = column, data = df, palette = 'viridis')
    plt.title(f"Univariate Analysis - {column}")
    plt.show()
```

```python
## zero means product not taken , 1 means product taken.
```

```python
cross_tab = pd.crosstab(df['MaritalStatus'], df['ProdTaken'], normalize = 'index')
cross_tab.plot(kind = 'bar', stacked=True, color = ['skyblue', 'lightcoral'], figsize = (10, 6))
plt.title("stacked bar blot of marital status and product taken")
plt.xlabel("Marital status")
plt.ylabel('percentage')
plt.show()
```

### insights : 
80% of divorced ppl taken product & 5% not taken

80 % of married ppl taken product & 5% not taken

70 % of  signle ppl taken product & 15 % not taken

78% of unmarried ppl taken product & 10 % not taken
