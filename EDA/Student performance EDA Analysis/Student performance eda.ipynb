```python
# urban std

sns.kdeplot(df.loc[df['address'] == 'U', 'G1'], label = 'Urban', shade=True)
```

```python
#do urban students score higher than rural student?


sns.kdeplot(df.loc[df['address'] == 'U', 'G1'], label = 'Urban', shade=True)
sns.kdeplot(df.loc[df['address'] == 'R', 'G1'], label = 'Rural', shade=True)
plt.title("performance comparison for G1")
plt.xlabel("Grade 1")
plt.show()
```

```python
## blue for urban 
## yellow for rural
## performance wise urban are more compare to rural
```

```python
## G2

sns.kdeplot(df.loc[df['address'] == 'U', 'G2'], label = 'Urban', shade=True)
sns.kdeplot(df.loc[df['address'] == 'R', 'G2'], label = 'Rural', shade=True)
plt.title("performance comparison for G2")
plt.xlabel("Grade 2")
plt.show()
```

```python
### same grade for urban & rural
```

```python
### G3
sns.kdeplot(df.loc[df['address'] == 'U', 'G3'], label = 'Urban', shade=True)
sns.kdeplot(df.loc[df['address'] == 'R', 'G3'], label = 'Rural', shade=True)
plt.title("performance comparison for G3")
plt.xlabel("Grade 3")
plt.show()
```
