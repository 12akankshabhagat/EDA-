```python
# Check if 'SMOKING' column exists
if 'SMOKING' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(x='SMOKING', hue='LUNG_CANCER', data=df, palette=['yellow' ,'green'])
    plt.title('Lung Cancer Distribution by Smoking Status')
    plt.xlabel('Smoking Status')
    plt.ylabel('Count')
    plt.show()
```

```python
df_smoking_causes_lung_cancer = df.groupby("SMOKING")["LUNG_CANCER"].value_counts().unstack()
df_smoking_causes_lung_cancer
```

```python
fig,ax =plt.subplots(1,2,figsize= (11,5))
ax[0].pie(df_smoking_causes_lung_cancer.iloc[0],labels=["Positive","Negative"],autopct="%1.1f%%",startangle=90)
ax[0].set_title('Lung Cancer Distribution Among non-Smokers')

ax[1].pie(df_smoking_causes_lung_cancer.iloc[1],labels=["Positive","Negative"],autopct="%1.1f%%",startangle=90)
ax[1].set_title('Lung Cancer Distribution Among Smokers')

plt.show()
```

Observation :

Smoking  is main Causes of Lung Cancer.

```python
# Create age bins
age_bins = [0, 20, 40, 60, 80, 100]
age_labels = ['0-20', '21-40', '41-60', '61-80', '80+']
df['AGE_GROUP'] = pd.cut(df['AGE'], bins=age_bins, labels=age_labels)

# Plot Lung Cancer distribution by Age Group
plt.figure(figsize=(10, 5))
sns.countplot(x='AGE_GROUP', hue='LUNG_CANCER', data=df, palette=['Blue', 'Pink'])
plt.title('Lung Cancer Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()

```

```python
# YELLOW_FINGERS count
plt.figure(figsize=(12,5))
ax =sns.countplot(x="YELLOW_FINGERS", data=df,hue="LUNG_CANCER")

for i in ax.containers:
    ax.bar_label(i)

plt.legend(["Does not have cancer","Has cancer"])
plt.show()
```

```python
plt.figure(figsize=(12,5))
ax =sns.countplot(x="CHEST_PAIN", data=df,hue="LUNG_CANCER")

for i in ax.containers:
    ax.bar_label(i)

plt.legend(["Does not have cancer","Has cancer"])
plt.show()
```

```python
# Histogram of numerical features
df.hist(figsize=(12, 8))
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```
